```python
# imports 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import mean_absolute_error


# import data
X = pd.read_csv('/kaggle/input/genetic-data-classification-practice-challenge/train.csv',header=None)
test_X = pd.read_csv('/kaggle/input/genetic-data-classification-practice-challenge/test_x.csv',header=None)
```

```python
X.head()
```

```python
# Check for missing or na values

cols_with_missing = [col for col in X.columns if X[col].isnull().any()]
cols_with_na = [col for col in X.columns if X[col].isna().any()]

print(cols_with_missing, cols_with_na)
```

```python
np.shape(X)
```

```python
# Pull out the target column

y = X[20531]
X.drop([20531], axis=1, inplace=True)
```

```python
X.head()
```

```python
# Break off validation set from training data

X_train, X_valid, y_train, y_valid = train_test_split(X, y,train_size=0.8, test_size=0.2,random_state=0)
```

```python
# Define a random forest model

rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)
rf_val_predictions = rf_model.predict(X_valid)
rf_val_mae = mean_absolute_error(rf_val_predictions, y_valid)

print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
model = RandomForestClassifier(random_state=1)
model.fit(X, y)
preds_test = model.predict(test_X)
preds_prob_test = model.predict_proba(test_X)
```

```python
np.shape(preds_prob_test)
```

```python
prob_matrix = np.asmatrix(preds_prob_test)

solutions = pd.concat([test_X, pd.DataFrame(prob_matrix)], axis=1)
```
